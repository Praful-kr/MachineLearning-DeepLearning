# **Apply linear regression with gradient descent**

**Data Description**: Here the dataset taken has 7 attributes and from that we can consider the last attribute as dependent variable and any other attribute as independent variable.

**There are total 1338 rows in the dataset**  

```
#Importing the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (28.5, 9.0)

# Preprocessing Input data
data = pd.read_csv('/insurance.csv')
data.head()
```

```
data.describe()
```

**Here we are considering only 2 columns(column 2 i.e.** **BMI and column 6 i.e. charges) for simplicity as we have to perform linear regression.**

**Column 2 is independent variable and Column 6 is dependent variable.** 

```
X = data.iloc[:, 2]
Y = data.iloc[:, 6]
plt.grid(True)
plt.scatter(X,Y)
plt.title("Data distribution")
plt.show()
```

# Test 1:

**Taking step size = 0.0001 and number of iterations = 10,** **we get the value** **of m = 382.04379968605264 and c = 12.075730798029474** 

```
# Building the model
m = 0
c = 0

L = 0.0001  # The learning Rate
epochs = 10  # The number of iterations to perform gradient descent

n = float(len(X)) # Number of elements in X

# Performing Gradient Descent 
for i in range(epochs): 
    Y_pred = m*X + c  # The current predicted value of Y
    D_m = (-2/n) * sum(X * (Y - Y_pred))  # Derivative wrt m
    D_c = (-2/n) * sum(Y - Y_pred)  # Derivative wrt c
    m = m - L * D_m  # Update m
    c = c - L * D_c  # Update c
print (m, c)
```

```
# Making predictions
Y_pred = m*X + c

plt.scatter(X, Y)
plt.grid(True)
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red') # predicted
plt.show()
```
