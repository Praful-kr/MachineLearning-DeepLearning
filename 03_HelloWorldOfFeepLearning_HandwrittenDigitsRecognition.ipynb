## **Importing the required libraries**

```
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline 
import numpy as np
```

## **Load train and test dataset from mnist**

```
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data() 
```

**number of rows in training data:**




```
len(X_train)
```

**number of rows in test data:**

```
len(X_test)
```

**dimensions of the data available:**

```
X_train[0].shape
```

**Grayscale image of the data:**

```
plt.matshow(X_train[5])
```

```
y_train[5]
```

```
X_train.shape
```

**perform scaling to improve accuracy of the neural network:**

```
X_train = X_train/255
X_test = X_test/255
```

## **TASK 0: Creating Neural Network**

**convert 2d matrix to 1d matrix in order to feed at the input layer:**

```
X_train_flattened = X_train.reshape(len(X_train), 28*28)
X_test_flattened = X_test.reshape(len(X_test), 28*28)
```

```
X_train_flattened.shape
```

```
X_test_flattened.shape
```

```
model = keras.Sequential([
  keras.layers.Dense(10, input_shape = (784,), activation = 'sigmoid')
])

model.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)

model.fit(X_train_flattened, y_train, epochs = 5)
```

```
model.evaluate(X_test_flattened, y_test)
```

## **TASK 1: Analysing the performance**

***Batch size = 32, Number of hidden layers = 1, Learning rate = 0.01 ***

```
model = keras.Sequential([
  keras.layers.Dense(10, input_shape = (784,), activation = 'sigmoid')
])

opt = keras.optimizers.Adam(learning_rate=0.01)
model.compile(
    optimizer = opt,
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)

history = model.fit(X_train_flattened, y_train, epochs = 5, batch_size = 32)
print(history.history.keys())
plt.plot(history.history['accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

***Batch size = 16, Number of hidden layers = 2, Learning rate = 0.05 ***

```
model = keras.Sequential([
  keras.layers.Dense(100, input_shape = (784,), activation = 'relu'),
  keras.layers.Dense(10, activation = 'sigmoid')
])

opt = keras.optimizers.Adam(learning_rate=0.05)
model.compile(
    optimizer = opt,
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)

history = model.fit(X_train_flattened, y_train, epochs = 5, batch_size = 16)
print(history.history.keys())
plt.plot(history.history['accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

***Batch size = 16, Number of hidden layers = 2, Learning rate = 0.01***

```
model = keras.Sequential([
  keras.layers.Dense(100, input_shape = (784,), activation = 'relu'),
  keras.layers.Dense(10, activation = 'sigmoid')
])

opt = keras.optimizers.Adam(learning_rate=0.01)
model.compile(
    optimizer = opt,
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)

history = model.fit(X_train_flattened, y_train, epochs = 5, batch_size = 16)
print(history.history.keys())
plt.plot(history.history['accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

### ***We observe that we get the maximum accuracy by reducing the learning rate, decreasing the batch size and limiting the number of hidden layers.***

***Running on test data:***

```
model.evaluate(X_test_flattened, y_test)
```

## **TASK 2: Confusion Matrix**

**We are converting the predicted values(decimal) to concrete values(as the ones in test labels) for confusion matrix prediction**

```
y_predicted_labels = [np.argmax(i) for i in y_predicted]
y_predicted_labels[:5]
```

```
y_test[:5]
```

***Confusion Matrix:***

```
cm = tf.math.confusion_matrix(labels = y_test, predictions = y_predicted_labels)
cm
```

```
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot = True, fmt = 'd')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

## **TASK 3: Overfitting**

**Training data = first 1000 of the 50000 of the training data**

**Batch size = 32**

**Epochs = 500** 

```
X_train_overfit = X_train_flattened[:1000]
train_labels_overfit = y_train[:1000]
model.fit(X_train_overfit, train_labels_overfit, batch_size = 32, epochs = 500)
```

**In 500 epochs we observe that the accuracy was perfect after some time and it was same after that.**
